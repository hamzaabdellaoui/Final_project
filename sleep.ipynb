About Dataset
This dataset is a synthetic but highly realistic sleep-tracking dataset generated to help students, researchers, data scientists, and AI engineers develop and test predictive models related to sleep quality, wellness, and human behavior.

It simulates 20,000 nightly sleep sessions recorded by 2,000 smartwatch users between 2018 and 2025, including physiological, environmental, and behavioral signals.

All data are fully synthetic and privacy-safe, making it ideal for educational use, research projects, and AI model experimentation without ethical or privacy concerns.

```python
x=df_scaled.drop('insomnia_flag',axis=1)
y=df_scaled['insomnia_flag']
```

```python
from sklearn.feature_selection import mutual_info_classif
mi=mutual_info_classif(x,y,random_state=42)
mi_series =pd.Series(mi,index=x.columns).sort_values(ascending=False)
mi_series
```

```python
threshold=0.00001
selected_features=mi_series[mi_series>threshold].index
X = [selected_features]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import  SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
df_scaled.shape
```

```python
x.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
svm_clf = SVC(kernel='rbf',C=1.0,random_state=42)
svm_clf.fit(x_train,y_train)
y_pred_svm = svm_clf.predict(x_test)

acc_svm = accuracy_score(y_test,y_pred_svm)
print (f" SVC  Accuracy: {acc_svm:.2f}")
print('Confusion matrix rbf \n', confusion_matrix(y_test,y_pred_svm))
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
```

```python


model = Sequential(
    [
     Dense(units=512,activation='relu',input_shape=(36,)),
     Dropout(0.1),
     Dense(units=128,activation='relu'),
     Dropout(0.2),
     Dense(units=256,activation='relu'),
     Dropout(0.3),
     Dense(64,activation='relu'),
     Dropout(0.2),
     Dense(32,activation='relu'),
     Dense(units=2,activation='sigmoid')
    ]
)

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python


early_stop = EarlyStopping(
    monitor='val_loss',       # ce qu'on surveille (val_loss est le plus courant)
    patience=15,               # nbre d'√©poques sans am√©lioration avant d'arr√™ter
    restore_best_weights=True,# remet les poids de la meilleure √©poque
    mode='min',               # 'min' car on veut minimiser la loss
    verbose=1
)
```

```python
model.fit(
    x_train,
    y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    callbacks=[early_stop]
)
```

```python
loss,accuracy = model.evaluate(x_test,y_test)
print(f'Loss: {loss} , Accuracy: {accuracy}')

y_test_pred = np.argmax(model.predict(x_test, verbose=1), axis=1)
confusion_matrix(y_test, y_test_pred)
```
